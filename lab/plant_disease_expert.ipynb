```python
import os
import random
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from PIL import Image

#configurando o caminho do dataset # a pasta image data base tava duplicada
DATASET_PATH = "../datasets/PlantDiseaseExpert/Image Data base/Image Data base"
```

```python
import os

todas_pastas = os.listdir(DATASET_PATH)

# classes que o artigo removeu 
classes_removidas = ["Nitrogen deficiency in plant", "Waterlogging in plant"]

# filtrando o dataset
classes_ativas = [cls for cls in todas_pastas if cls not in classes_removidas]
classes_ativas = sorted(classes_ativas)

print(f"Total de classes na pasta: {len(todas_pastas)}")
print(f"Classes removidas: {classes_removidas}")
print(f"Total de classes usadas para o experimento: {len(classes_ativas)}")
print("-" * 60)
print("Amostra das primeiras 10 classes ativas:")
for i, cls in enumerate(classes_ativas[:10], 1):
    print(f"  {i}. {cls}")
```

```python
#mostrando exemplos
def plot_dataset_examples(DATASET_PATH, classes_ativas, num_examples=4):
    fig, axes = plt.subplots(1, num_examples, figsize=(16, 4))
    
    if num_examples == 1:
        axes = [axes]
        
    for i in range(num_examples):
        #seleciona uma classe ativa 
        random_cls = random.choice(classes_ativas)
        cls_dir = os.path.join(DATASET_PATH, random_cls)
        
        #seleciona uma imagem aleatória 
        random_img = random.choice(os.listdir(cls_dir))
        img_path = os.path.join(cls_dir, random_img)
        
        
        img = Image.open(img_path)
        axes[i].imshow(img)
        axes[i].set_title(random_cls.replace("___", "\n").replace("__", "\n"), fontsize=9)
        axes[i].axis('off')
        
    plt.tight_layout()
    plt.show()
plot_dataset_examples(DATASET_PATH, classes_ativas, num_examples=4)
```
